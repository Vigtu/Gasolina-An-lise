### 4. Você deve ter percebido que as colunas DATA INICIAL e DATA FINAL estão formatadas como string. Utilizando o método pd.to_datetime(), converta ambas colunas para Timestamp / Datetime.

```python
# Verificar se há dados ausentes na coluna 'DATA INICIAL'
dados_ausentes_inicial = df['DATA INICIAL'].isnull().sum()
print("Número de dados ausentes na coluna 'DATA INICIAL':", dados_ausentes_inicial)

# Verificar se há dados ausentes na coluna 'DATA FINAL'
dados_ausentes_final = df['DATA FINAL'].isnull().sum()
print("Número de dados ausentes na coluna 'DATA FINAL':", dados_ausentes_final)
```

```python
df['DATA INICIAL'] = pd.to_datetime(df['DATA INICIAL'], format ='%Y-%m-%d')
df['DATA FINAL'] = pd.to_datetime(df['DATA FINAL'], format ='%Y-%m-%d')
```
